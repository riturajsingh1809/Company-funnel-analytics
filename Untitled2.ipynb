```python
# ==========================================
# COMPANY FUNNEL ANALYSIS PROJECT
# ==========================================

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_excel("/content/company_funnel_dataset.xlsx")

print("="*50)
print("DATASET LOADED SUCCESSFULLY")
print("="*50)

# First 5 Rows
print("\nFirst 5 Rows")
print(df.head())

# Dataset Information
print("\nDataset Information")
df.info()

# Shape
print("\nDataset Shape")
print(df.shape)

# Missing Values
print("\nMissing Values")
print(df.isnull().sum())

# Duplicate Values
print("\nDuplicate Values")
print(df.duplicated().sum())

# Statistical Summary
print("\nStatistical Summary")
print(df.describe())

# ==========================================
# COMPANY FUNNEL KPIs
# ==========================================

total_students = len(df)
viewed = df["Viewed"].sum()
applied = df["Applied"].sum()
shortlisted = df["Shortlisted"].sum()
hired = df["Hired"].sum()

print("\n")
print("="*50)
print("COMPANY FUNNEL KPI")
print("="*50)

print("Total Students :", total_students)
print("Jobs Viewed :", viewed)
print("Applications :", applied)
print("Shortlisted :", shortlisted)
print("Hired :", hired)

# ==========================================
# FUNNEL CHART
# ==========================================

stages = [
    "Students",
    "Viewed",
    "Applied",
    "Shortlisted",
    "Hired"
]

values = [
    total_students,
    viewed,
    applied,
    shortlisted,
    hired
]

plt.figure(figsize=(8,5))
plt.bar(stages, values)
plt.title("Company Funnel")
plt.xlabel("Stages")
plt.ylabel("Count")
plt.grid(axis='y')

plt.savefig("company_funnel_chart.png")

plt.show()

# ==========================================
# MATCH SCORE DISTRIBUTION
# ==========================================

plt.figure(figsize=(8,5))
plt.hist(df["Match_Score"], bins=10)

plt.title("Match Score Distribution")
plt.xlabel("Match Score")
plt.ylabel("Students")

plt.savefig("match_score_distribution.png")

plt.show()

# ==========================================
# JOB CATEGORY ANALYSIS
# ==========================================

category = df["Job_Category"].value_counts()

plt.figure(figsize=(8,5))
category.plot(kind="bar")

plt.title("Job Category Distribution")
plt.xlabel("Category")
plt.ylabel("Count")

plt.savefig("job_category_distribution.png")

plt.show()

# ==========================================
# FINAL MESSAGE
# ==========================================

print("\n")
print("="*60)
print("PROJECT COMPLETED SUCCESSFULLY")
print("="*60)
```

```python
# ==========================================
# FINAL PROJECT OUTPUT & REPORT FILES
# ==========================================

import pandas as pd

# KPI Summary
kpi_summary = pd.DataFrame({
    "Metric": [
        "Total Students",
        "Jobs Viewed",
        "Applications",
        "Shortlisted",
        "Hired"
    ],
    "Value": [
        total_students,
        viewed,
        applied,
        shortlisted,
        hired
    ]
})

print("\nKPI SUMMARY")
print(kpi_summary)

# Save KPI Summary
kpi_summary.to_csv("kpi_summary.csv", index=False)

# Job Category Summary
category_summary = (
    df.groupby("Job_Category")
      .agg(
          Total_Students=("Student_ID", "count"),
          Average_Match_Score=("Match_Score", "mean"),
          Applications=("Applied", "sum"),
          Shortlisted=("Shortlisted", "sum"),
          Hired=("Hired", "sum")
      )
      .reset_index()
)

print("\nJOB CATEGORY SUMMARY")
print(category_summary)

category_summary.to_csv("job_category_summary.csv", index=False)

# Funnel Summary
funnel_summary = pd.DataFrame({
    "Stage": [
        "Students",
        "Viewed",
        "Applied",
        "Shortlisted",
        "Hired"
    ],
    "Count": [
        total_students,
        viewed,
        applied,
        shortlisted,
        hired
    ]
})

print("\nFUNNEL SUMMARY")
print(funnel_summary)

funnel_summary.to_csv("company_funnel_summary.csv", index=False)

print("\nAll report files created successfully.")
print("Generated files:")
print("- company_funnel_chart.png")
print("- match_score_distribution.png")
print("- job_category_distribution.png")
print("- kpi_summary.csv")
print("- job_category_summary.csv")
print("- company_funnel_summary.csv")
```
